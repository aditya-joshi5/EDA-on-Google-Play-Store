```python
def extract_word(item):
  if str(item)[-1]=='M':  # when last word is 'M'
    res=item[:-1]         # Take upto second last term
    res=float(res)*1024   # conversion of mb to kb
    return res
  elif str(item)[-1]=='k':
    res=item[:-1]
    return res
  else:
    return str(np.nan)
```

```python
extract_word('99M') # Check--extract_word function working properly
```

```python
df_copy.Size=df_copy.Size.apply(extract_word) # apply extract_word function to the size column
```

```python
def remove_sym (num):
  if str(num)[-1] =='+':
    res=str(num)[:-1]
    return res
  else:
    num=str(num)
    return num
```
